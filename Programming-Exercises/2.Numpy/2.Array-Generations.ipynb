# Section-5: Sampling normal distribution using **np.random.normal** 

## Examples

```python
from scipy.stats import norm

# Declare variables
mu, sigma = 5, 3
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = norm.pdf(x, mu, sigma)


plt.figure(figsize=(6,4))
plt.plot(x, y, c='r', linewidth=2)
title = f"Normal distribution with mu={mu}, sigma={sigma}"
plt.title(title)
plt.show()
```

```python
x = np.linspace(0, 5, 100)
y1 = 2*x + 1
y2 = -5*x + 3

plt.figure(figsize=(9, 4), layout='constrained')

ax = plt.subplot(1,2, 1) # 1,2,1 = 1 row, 2 cols, activate subplot-1 (i.e., left)
ax.plot(x, y1, c='b')
ax.set_title("$y = 2x + 1$")

ax = plt.subplot(1,2, 2) # 1,2,2 = 1 row, 2 cols, activate subplot-2 (i.e., right)
ax.plot(x, y2, c='r')
ax.set_title("$y = -5x + 3$")
plt.suptitle("Some functions")
plt.show()
```

## Q.2- Sampling 2D-normal

### Q.2.1 Examples

```python
# Declare variables
mu = (5, 3)
sigma = (3,4)
N = 10000

# Generate data from 2D-normal(mu sigma)
X = np.random.normal(mu, sigma, (N,2)) #N rows, 2 cols

# Show the 2D-histogram
fig = plt.figure(figsize=(6,4))
ax = fig.gca() # you can get axes from plt.subplot(,,)
_, _, _, im = ax.hist2d(X[:, 0], X[:, 1], bins=10*10, density=True)
title = "Histogram from {} points".format(N)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title(title)
fig.colorbar(im, ax=ax)
plt.show()
```
